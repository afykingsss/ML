```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df=pd.read_csv('Housing.csv')
df
```

```python
main_road=pd.get_dummies(df['mainroad'],drop_first= True)
guest_room=pd.get_dummies(df['guestroom'],drop_first=True)
basement=pd.get_dummies(df['basement'],drop_first=True)
hotwaterheating=pd.get_dummies(df['hotwaterheating'],drop_first=True)
air_conditioning=pd.get_dummies(df['airconditioning'],drop_first=True)
prefarea=pd.get_dummies(df['prefarea'],drop_first=True)
furnishing_status=pd.get_dummies(df['furnishingstatus'],drop_first=True)
df
```

```python
df=pd.concat([df,main_road],axis=1)
df=df.rename(columns={"yes":"mainroad_yes"})
df=pd.concat([df,guest_room],axis=1)
df=df.rename(columns={"yes":"guestroom_yes"})
df=pd.concat([df,basement],axis=1)
df=df.rename(columns={"yes":"basement_yes"})
df=pd.concat([df,hotwaterheating],axis=1)
df=df.rename(columns={"yes":"hotwaterheating_yes"})
df=pd.concat([df,air_conditioning],axis=1)
df=df.rename(columns={"yes":"airconditioning_yes"})
df=pd.concat([df,prefarea],axis=1)
df=df.rename(columns={"yes":"prefarea_yes"})
df=pd.concat([df,furnishing_status],axis=1)
df = df.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'],axis=1)
df
```

```python
X=df.iloc[:, 1:]
y=df.iloc[:, 0]
```

```python
bestfeatures=SelectKBest(score_func=chi2,k=10)
fit=bestfeatures.fit(X,y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
best_features = featureScores.nlargest(14, 'Score')
best_features
```

```python
print(df.isnull().sum())
df = df.drop_duplicates()
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
hasil = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
hasil
```

```python
print('Model R^2 Square value', metrics.r2_score(y_test, y_pred))
```
