```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
import numpy as np
```

```python
df=pd.read_csv("apple_quality.csv")
df = df.drop(['A_id'], axis=1)
df
```

```python
scaler= MinMaxScaler()
scaled=scaler.fit_transform(df[['Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity']])

scaled_df = pd.DataFrame(scaled, columns=('Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity'))
df[['Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity']] = scaled_df



```

```python
X = df.iloc[:,0:7]  #independent columns
y = df.iloc[:,-1]   #target column quality
```

```python
bestfeatures=SelectKBest(score_func=chi2,k=7)
fit=bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
```

```python
featureScores
```

```python
df=df.drop(['Weight','Acidity'],axis=1)
```

```python
df #setelah Cleaning Data
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 5].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model = RandomForestClassifier(criterion = "gini", n_estimators = 10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print(model.feature_importances_)
```
